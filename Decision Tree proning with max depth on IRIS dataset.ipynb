```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
import numpy as np
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
#independent feature and dependent features
X=df.iloc[:,:-1]
y=df.iloc[:,4]
```

```python
X
```

```python
y.unique()
```

```python
y.replace(['setosa', 'versicolor', 'virginica'],[0,1,2],inplace=True)
```

```python
y
```

```python
### train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treemodel=DecisionTreeClassifier()
```

```python
treemodel.fit(X_train,y_train)
```

```python
#prediction
y_pred=treemodel.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
treemodel1=DecisionTreeClassifier(max_depth=2)
```

```python
treemodel1.fit(X_train,y_train)
```

```python
#prediction
y_pred1=treemodel1.predict(X_test)
```

```python
y_pred1
```

```python
score1=accuracy_score(y_pred1,y_test)
print(score1)
```

```python
print(classification_report(y_pred1,y_test))
```
